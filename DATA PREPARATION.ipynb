```python
import pandas as pd
import pandas_ta as ta
```

```python
train = pd.read_csv('train_data.csv',
                parse_dates = ["date"],
                index_col = ["date"])
labels = pd.read_csv(r"train_target.csv")
test = pd.read_csv("test_data.csv")
```

# CALCULATING THE FOLLOWING TIs:
* MACD
* KAMA
* STOCHASTIC OSCILLATOR
* AROON INDICATOR
* PARABOLIC STOP AND REVERSE(SAR)

### MACD

```python
macd_indicator = train.ta.macd(close='close', fast=12, slow=26, signal=9, append=False)
test_macd_indicator = test.ta.macd(close='close', fast=12, slow=26, signal=9, append=False)
```

```python
macd_indicator.tail(30)
```

### KAMA

```python
kama_indicator = ta.kama(train["close"].squeeze(),length = 10)
test_kama_indicator = ta.kama(test["close"].squeeze(),length = 10)
```

```python
kama_indicator = kama_indicator.to_frame()
test_kama_indicator = test_kama_indicator.to_frame()
kama_indicator
```

### STOCHASTIC OSCILLATOR

```python
stochastic_oscillator = train.ta.stoch( window=14 , 
                                        smooth_window = 3, 
                                        fillna = False)
test_stochastic_oscillator = test.ta.stoch( window=14 , 
                                        smooth_window = 3, 
                                        fillna = False)
```

```python
stochastic_oscillator
```

### AROON INDICATOR

```python
aroon_indicator = train.ta.aroon(window = 25, fillna = False)
test_aroon_indicator = test.ta.aroon(window = 25, fillna = False)
aroon_indicator  
```

### SAR INDICATOR

```python
sar_indicator = train.ta.psar(step = 0.02, max_step = 0.2, fillna = False)
test_sar_indicator = test.ta.psar(step = 0.02, max_step = 0.2, fillna = False)
sar_indicator
```

### MERGING ALL THE INDICATORS 

```python
train = train.merge(sar_indicator,left_index=True, right_index=True)
train = train.merge(aroon_indicator,left_index=True, right_index=True)
train = train.merge(stochastic_oscillator,left_index=True, right_index=True)
train = train.merge(macd_indicator,left_index=True, right_index=True)
train = train.merge(kama_indicator,left_index=True, right_index=True)

test = test.merge(sar_indicator,left_index=True, right_index=True)
test = test.merge(aroon_indicator,left_index=True, right_index=True)
test = test.merge(stochastic_oscillator,left_index=True, right_index=True)
test = test.merge(macd_indicator,left_index=True, right_index=True)
test = test.merge(kama_indicator,left_index=True, right_index=True)
train.head()
```
